```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

# import image

```python
gif_path = "p5_image.gif"
image = Image.open(gif_path)  #

image_array = np.array(image)  
print(image_array.shape)

plt.imshow(image_array, cmap='gray')

```

# perform SVD for image compression

```python
k = 800

U, s, Vh = np.linalg.svd(image_array)
Uk = U[:, :k]
sk = s[:k]
Vk = Vh[:k, :]
image_array_k = Uk @ np.diag(sk) @ Vk

image_array_k = (image_array_k - np.min(image_array_k)) / (np.max(image_array_k) - np.min(image_array_k))


plt.imshow(image_array_k, cmap='gray')

```
